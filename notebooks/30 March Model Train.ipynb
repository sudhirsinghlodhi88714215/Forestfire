```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pd.read_csv('Fire.csv')
```

```python
da=pd.read_csv('Fire.csv')
```

```python
da.head()
```

```python
da.info()
```

```python
X=da[['Temperature','RH','Ws','Rain','FFMC','DMC','DC','ISI','BUI','FWI','Classes','Regsion']]
```

```python
X
```

```python
Y=da['FWI']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(X,Y,test_size=0.30,random_state=30)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train.corr()
```

```python
plt.figure(figsize=(12,10))
corr=x_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def correlation(dataset,threshold):
    col_corr= set()
    corr_matrix= dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range (i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr      
```

```python
corr_feature=correlation(x_train,0.85)
```

```python
corr_feature
```

```python
x_train.drop(corr_feature,axis=1,inplace=True)
x_test.drop(corr_feature,axis=1,inplace=True)
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_sc=scaler.fit_transform(x_train)
x_test_sc=scaler.transform(x_test)
```

```python
x_train_sc
```

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=x_train)
plt.title('x_train')
plt.subplot(1,2,2)
sns.boxplot(data=x_train_sc)
plt.title('x_train_sc')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import  r2_score
linear=LinearRegression()
linear.fit(x_train_sc,y_train)
x_pred=linear.predict(x_test_sc)
mae=mean_absolute_error(y_test,x_pred)
score=r2_score(y_test,x_pred)
print( 'mean_absolute_error',mae)
print('r2 score',score)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import  r2_score
ridge=Ridge()
ridge.fit(x_train_sc,y_train)
x_pred=ridge.predict(x_test_sc)
mae=mean_absolute_error(y_test,x_pred) 
score=r2_score(y_test,x_pred)
print( 'mean_absolute_error',mae)
print('r2 score',score)
```
